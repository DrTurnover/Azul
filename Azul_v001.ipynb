```python
#Вспомогательный метод конвертации представления плиток
def convert_tiles_list(tiles_raw):
    tiles_digital = []
    colors = ['RED', 'BLUE', 'WHITE', 'BLACK', 'YELLOW']
    for color in colors:
        if tiles_raw.count(color) > 0:
            tiles_digital.append([color, tiles_raw.count(color)])

    return tiles_digital

setattr(Game, 'convert_tiles_list', convert_tiles_list)

def is_finished(self):
    for player_num in self.queue:
        if self.players[player_num].is_finished():
            Locale.message(self, 'player_finished_game', player_num)
            return True
    return False

setattr(Game, 'is_finished', is_finished)
```

Класс Table задает сумку с плитками, фабрики и кучу фишек в середине стола.
Куча построена как фабрика с номером 0

```python
class Table:
    def __init__(self, num_players, game_style):
        self.bag = ['RED', 'BLUE', 'WHITE', 'BLACK', 'YELLOW'] * 20
        self.used_tiles_bag = []
        
        self.num_players = num_players
        self.game_style = game_style 
        
        if num_players == 2:
            self.factories = {i: Game.empty_dict() for i in range(0, 6)}
        elif num_players == 3:
            self.factories = {i: Game.empty_dict() for i in range(0, 8)}
        else:
            self.factories = {i: Game.empty_dict() for i in range(0, 10)}
        
        self.starting_marker_on_table = True
    
    def convert(tiles_raw):
        '''
        Конвертирует список фишек в словарь с количеством
        Пример: ['RED', 'RED', 'BLACK'] -> {'RED': 2, 'BLUE': 0, 'WHITE': 0, 'BLACK': 1, 'YELLOW': 0}
        '''
        tiles_digital = {}
        colors = ['RED', 'BLUE', 'WHITE', 'BLACK', 'YELLOW']
        for color in colors:
            tiles_digital[color] = tiles_raw.count(color)
        return tiles_digital
     
    def fill_factories(self):
                
        random.shuffle(self.bag)
                
        for factory_num in self.factories.keys() - {0}:
            if len(self.bag) >= 4:
                tiles_to_serve = self.bag[:4]
                self.bag = self.bag[4:]
            else:
                self.bag = self.bag + self.used_tiles_bag
                self.used_tiles_bag = []
                random.shuffle(self.bag)
                if len(self.bag) >= 4:
                    tiles_to_serve = self.bag[:4]
                    self.bag = self.bag[4:]
                else:
                    tiles_to_serve = self.bag
                    self.bag = []
            self.add_tiles(factory_num, Table.convert(tiles_to_serve))
        
    def show(self):
        print('\n')
        print('----------------------------------------------------')
        print('Показ стола')
        for factory_num in self.factories.keys() - {0}:
            print('Фабрика ', factory_num, self.factories[factory_num])
        if self.starting_marker_on_table:
            print('Куча:', '[Стартовый маркер]', self.factories[0])
        else:
            print('Куча:', self.factories[0])
        print('----------------------------------------------------')
    
    def show_factory(self, factory_num):
        if factory_num == 0:
            print('Куча:', self.factories[factory_num])
        else:
            print('Фабрика', factory_num, self.factories[factory_num])
     
    def convert_2(tiles_dict):
        l = []
        for k, v in zip(tiles_dict.keys(), tiles_dict.values()):
            l += [k] * v
        return l
    
    def add_tiles(self, f_num, to_add):
        for k in self.factories[f_num].keys():
            self.factories[f_num][k] += to_add[k]
    
    def throw_used(self, used_tiles):
        self.used_tiles_bag += used_tiles
    
```

```python
def is_table_empty(self):
    #print('LABE', self.factories)
    if any(sum(f.values()) != 0 for f in self.factories.values()):
        return False
    else:
        return True
    
setattr(Table, 'is_table_empty', is_table_empty)        

def is_factory_empty(self, f_num):
    if sum(self.factories[f_num].values()) == 0:
        return True
    else:
        return False
    
setattr(Table, 'is_factory_empty', is_factory_empty)   

def is_color_present(self, f_num, color):
    return self.factories[f_num][color] != 0

setattr(Table, 'is_color_present', is_color_present)   
```

```python
def player_took_tiles(self, f_num, color):
    self.factories[f_num][color] = 0
    
    if f_num != 0:
        self.add_tiles(0, self.factories[f_num]) #скидываем оставшиеся плитки с завода в кучу
        self.factories[f_num] = Game.empty_dict()
    
setattr(Table, 'player_took_tiles', player_took_tiles)
```

Класс игрока
collection - планшет с плитками с левой стороны;
board_tiles - планшет с правой стороны, показывает, какие ячейки уже заполнены;
board_template - справочник по расположению цветов;
reverse_template - обратный справочник, по цвету дает позицию.
Последние два объекта неизменны, но экземпляр у каждого игрока. Надо бы сделать так, чтобы был только один экзепляр у всех. Решение - засунуть в класс Game, но потом

```python
class Player:
    def __init__(self, game_style):
        self.collection = list({'size': i, 'color': None, 'qty': 0} for i in range(1,6))
        self.board_tiles = np.full((5, 5), False)
        self.board_template = np.array((['RED', 'BLUE', 'WHITE', 'BLACK', 'YELLOW'], ['YELLOW', 'RED', 'BLUE', 'WHITE', 'BLACK'],\
                                  ['BLACK', 'YELLOW', 'RED', 'BLUE', 'WHITE'], ['WHITE', 'BLACK', 'YELLOW', 'RED', 'BLUE'],\
                                  ['BLUE', 'WHITE', 'BLACK', 'YELLOW', 'RED']))
        self.reverse_template = [{'RED': (0,0), 'BLUE': (0,1), 'WHITE': (0,2), 'BLACK': (0,3), 'YELLOW': (0,4)},\
                                 {'RED': (1,1), 'BLUE': (1,2), 'WHITE': (1,3), 'BLACK': (1,4), 'YELLOW': (1,0)},\
                                 {'RED': (2,2), 'BLUE': (2,3), 'WHITE': (2,4), 'BLACK': (2,0), 'YELLOW': (2,1)},\
                                 {'RED': (3,3), 'BLUE': (3,4), 'WHITE': (3,0), 'BLACK': (3,1), 'YELLOW': (3,2)},\
                                 {'RED': (4,4), 'BLUE': (4,0), 'WHITE': (4,1), 'BLACK': (4,2), 'YELLOW': (4,3)}]
        
        self.floor = {'RED': 0, 'BLUE': 0, 'WHITE': 0, 'BLACK': 0, 'YELLOW': 0}
        self.first_flag = False
        self.game_style = game_style
        self.score = 0
        
    def show(self):
        print('----------------------------------------------------')
        print('Очки:', self.score)
        print('Плитки:')
        
        for i in range(0, 5):
            #sub_1 = 'Линия номер' + ' ' + str(x['size']) +  str('  x
            sub_1 = str(i+1) + ') ' + str(self.collection[i])
            sub_2 = '\t\t'
            sub_3 = ''
            for k in range(0,5):
                if self.board_tiles[(i, k)]:
                    sub_3 += '[' + str(self.board_template[i, k]) + ']' + '\t'
                else:
                    sub_3 += str(self.board_template[i, k]) + '\t'
            print(sub_1 + sub_2 + sub_3)
            
        '''
        for x in self.collection:
            print('Линия номер', x['size'], x)
        print('Планшет:')
        print(self.board_tiles)
        print(self.board_template)
        '''
        if self.first_flag:
            print('Пол:', '[Стартовый маркер]', self.floor)
        else:
            print('Пол:', self.floor)
```

Покупка плиток с фабрик - игрок выбирает завод. 
Если он не пустой, выбирает цвет. 
Если такой цвет есть, то выбирает, на какую линию поставить плитки.
Если линиия не полна, подходит по цвету и на правом планшете цвет не заполнен - кладет на линию.
Если взял больше - скидывает остаток на пол

```python
def buy_tiles(self, table):
        print('Предложения фабрик:')
        table.show()
        #выбрать фабрику, с которой взять плитки
        f_num = self.choose_factory(table)
        if not f_num:
            print('Игрок отказался от игры')
        
        #выбрать цвет
        color = self.choose_color(table, f_num)
        if not color:
            print('Игрок отказался от игры')
        if color == '-1': #вернуться к выбору завода
            self.buy_tiles(table)
            return
        
        qty = table.factories[f_num][color]
        table.player_took_tiles(f_num, color) #удаление плиток
        
        if f_num == 0:
            self.take_start_marker(table)
        
        self.place_tiles_on_board(color, qty) #размещение плитки
                
setattr(Player, 'buy_tiles', buy_tiles)

def take_start_marker(self, table):
    table.starting_marker_on_table = False
    self.first_flag = True

setattr(Player, 'take_start_marker', take_start_marker)
```

```python
def choose_factory(self, table):
    print('Ваш планшет:')
    self.show()
    while True:
        print('Выберите номер фабрики, чтобы купить плитку. Если желаете поднять из кучи, введите 0 (ноль)')
        f_num = Game.input(self.game_style, 'num')
        f_num = Locale.check_name(f_num, 'f_num', optional = len(table.factories))
        if f_num is None:
            continue

        if table.is_factory_empty(f_num):
            if f_num == 0:
                print('В куче пока нет плиток, выберите завод')
            else:
                print('Завод пуст, выберите другой завод или кучу')
        else:
            break

    if f_num == 0:
        print('Вы выбрали подобрать из кучи. В ней доступны следующие плитки:')
    else:
        print('Вы выбрали фабрику', f_num, 'На ней доступны следующие плитки:')
    table.show_factory(f_num)
    return f_num

def choose_color(self, table, f_num):  
    while True:
        print('Выберите цвет плитки. Чтобы сменить фабрику, введите -1')
        color = Game.input(self.game_style)
        color = Locale.check_name(color, 'color')
        
        if not color:
            continue
        if color == '-1':
            print('Вы вернулись на предыдущий шаг - выбор фабрики')
            return color
        
        if table.is_color_present(f_num, color):
            break
        else:
            print('Такого цвета на заводе нет, выберите другой')
    print('Вы выбрали цвет плитки', color, 'Таких плиток в наличии', table.factories[f_num][color])
    return color                                    
```

```python
def place_tiles_on_board(self, color, qty):
    print('Ваш планшет:')
    self.show()
    while True:
        print('Выберите, на какую линию ставить фишки. Если хотите скинуть на пол, введите 0')
        line_num = Game.input(self.game_style)     
        
        line_num = Locale.check_name(line_num, 'line_num')
        if line_num is None:
            continue
        if line_num == -1:
            self.floor[color] += qty
            return
        
        color_position = self.reverse_template[line_num][color]

        if (self.collection[line_num]['color'] is None #линия слева пуста\ 
            or self.collection[line_num]['color'] == color and\
            self.collection[line_num]['qty'] < self.collection[line_num]['size'])\
        and not self.board_tiles[color_position]:
            break
        else:
            print('Нельзя разместить на этой линии плитки. Выберите другую')

    if self.collection[line_num]['color'] is None:
        self.collection[line_num]['color'] = color
    if self.collection[line_num]['qty'] + qty <=  self.collection[line_num]['size']:
        self.collection[line_num]['qty'] =+ qty
    else:
        self.floor[color] += qty - (self.collection[line_num]['size'] - self.collection[line_num]['qty'])
        self.collection[line_num]['qty'] = self.collection[line_num]['size']

    self.show()
    
```

```python
def scores(self, table):
    self.show()
    add_score = 0
    for line_num in range(0, 5):
        if self.collection[line_num]['size'] == self.collection[line_num]['qty']:
            color = self.collection[line_num]['color']
            print('Мостим плитку с линии', line_num, 'цвета', color)
            position = self.reverse_template[line_num][color]
            self.board_tiles[position] = True
            used_tiles = Table.convert_2({color: self.collection[line_num]['qty'] - 1})
            self.collection[line_num]['qty'] = 0
            self.collection[line_num]['color'] = None
            table.throw_used(used_tiles)
            
            add_score += 1
            for vert_above in range(position[0] - 1, -1, -1):
                if self.board_tiles[vert_above, position[1]]:
                    add_score += 1
                else:
                    break
            for vert_below in range(position[0] + 1, 5, 1):
                if self.board_tiles[vert_below, position[1]]:
                    add_score += 1
                else:
                    break
            for hor_left in range(position[1] - 1, -1, -1):
                if self.board_tiles[position[0], hor_left]:
                    add_score += 1
                else:
                    break
            for hor_right in range(position[1] + 1, 5, 1):
                if self.board_tiles[position[0], hor_right]:
                    add_score += 1
                else:
                    break
        print('Очков начислено:', add_score)
        self.score += add_score
        
    minus_score = 0
    minus_count = 0

    if self.first_flag:
        minus_count += 1

    minus_count += sum(self.floor.values())

    if minus_count == 0:
        pass
    elif minus_count == 1:
        minus_score += 1
    elif minus_count == 2:
        minus_score += 2
    elif minus_count == 3:
        minus_score += 4
    elif minus_count == 4:
        minus_score += 6
    elif minus_count == 5:
        minus_score += 8
    elif minus_count == 6:
        minus_score += 11
    elif minus_count >= 7:
        minus_score += 14

    print('Штрафных очков начислено:', minus_score)

    self.score -= minus_score
    self.score = max(self.score, 0)

    used_tiles = Table.convert_2(self.floor)
    self.floor = Game.empty_dict()
    table.throw_used(used_tiles)

        
        
    return max(self.score, 0)

def final_scores(self):
    column_score = 0
    for x in np.sum(self.board_tiles, 0):
        if x == 5:
            column_score += 7
    row_score = 0
    for x in np.sum(self.board_tiles, 1):
        if x == 5:
            column_score += 2
    
    color_score = 0
    if self.board_tiles[(0,0)] == self.board_tiles[(1,1)] == self.board_tiles[(2,2)] == self.board_tiles[(3,3)]\
    == self.board_tiles[(4,4)] == True:
        color_score += 10
    if self.board_tiles[(0,1)] == self.board_tiles[(1,2)] == self.board_tiles[(2,3)] == self.board_tiles[(3,4)]\
    == self.board_tiles[(4,0)] == True:
        color_score += 10
    if self.board_tiles[(0,2)] == self.board_tiles[(1,3)] == self.board_tiles[(2,4)] == self.board_tiles[(3,0)]\
    == self.board_tiles[(4,1)] == True:
        color_score += 10
    if self.board_tiles[(0,3)] == self.board_tiles[(1,4)] == self.board_tiles[(2,0)] == self.board_tiles[(3,1)]\
    == self.board_tiles[(4,2)] == True:
        color_score += 10
    if self.board_tiles[(0,4)] == self.board_tiles[(1,0)] == self.board_tiles[(2,1)] == self.board_tiles[(3,2)]\
    == self.board_tiles[(4,3)] == True:
        color_score += 10
        
    print('Бонусные очки:', column_score, row_score, color_score)
    self.score += column_score + row_score + color_score
    return self.score

def is_finished(self):
    if any(list(x == 5 for x in np.sum(self.board_tiles, 1))):
        return True
    else:
        return False
```

```python
class Locale():
    def check_name(string_to_check, string_type, optional = None):
        if string_type == 'color':
            if string_to_check in ['RED', 'BLUE', 'WHITE', 'BLACK', 'YELLOW', '-1']:
                return string_to_check
            else:
                print('Некорректный цвет')
                return None
        elif string_type == 'f_num':
            try:
                if 0 <= (int(string_to_check)) <= optional - 1:
                    return int(string_to_check)
                else:
                    print('Некорректный номер фабрики')
                    return None
            except:
                print('Некорректный номер фабрики')
                return None
        elif string_type == 'line_num':
            try:
                if 0 <= (int(string_to_check)) <= 5:
                    return int(string_to_check) - 1
                else:
                    print('Некорректный номер линии')
                    return None
            except:
                print('Некорректный номер фабрики')
                return None
        else:
            print('Unknown input')
            return None
    def message(game, message_name, optional = None):
        if message_name == 'game_started':
            print('\n')
            print('****************************************************')
            print('Игра началась!')
            print('Количество игроков:', game.num_players, 'Стиль игры:', game.game_style)
            print('****************************************************')
            print('\n')
        if message_name == 'factory_stage':
            print('\n')
            print('****************************************************')
            print('Стадия закупок с фабрик')
            print('****************************************************')
            print('\n')
        if message_name == 'factory_produced':
            print('Заводы произвели плитку')
        if message_name == 'player_turn':
            print('\n')
            print('Ход игрока', optional)
            print('----------------------------------------------------')
        if message_name == 'table_is_empty':
            print('\n')
            print('****************************************************')
            print('Стол пуст. Стадия закупок с фабрик завершена')
            print('****************************************************')
            print('\n')
        if message_name == 'scoring_stage':
            print('\n')
            print('****************************************************')
            print('Стадия подсчета очков')
            print('****************************************************')
            print('\n')
        if message_name == 'scoring_player_num':
            print('Подсчет очков игрока', optional)
            print('----------------------------------------------------')
        if message_name == 'scoring_stage_finished':
            print('\n')
            print('****************************************************')
            print('Стадия подсчета очков завершена')
            print('****************************************************')
            print('\n')
        if message_name == 'end_game':
            print('\n')
            print('****************************************************')
            print('Игра завершена')
            print('****************************************************')
            print('\n')
        if message_name == 'end_game_no_tiles':
            print('\n')
            print('****************************************************')
            print('Игра завершена: нет больше свободных плиток')
            print('****************************************************')
            print('\n')
        if message_name == 'end_game_no_tiles':
            print('\n')
            print('Игрок', optional, 'завершил игру')
            print('\n')
            
              
```
